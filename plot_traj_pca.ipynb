```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/git/cyanines/data/TRAJ_PCA/cis.csv')

ids = df['traj_id'].unique() # don't worry, this doesn't sort!

trajids = df['traj_id'].values
snips = (trajids[1:] - trajids[:-1]).nonzero()[0]
snips = np.append(0, snips)
snips = np.append(snips, -1)
slices = [slice(s+1, e) for s, e in zip(snips[:-1], snips[1:])] 

x = df['x']
y = df['y']
lines = [plt.plot(x[slice_], y[slice_], label=f'Traj. {ids[i]}')
         for i, slice_ in enumerate(slices)]

plt.legend()
plt.xlabel('PC1')
plt.ylabel('PC2')
```
